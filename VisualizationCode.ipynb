# **Sample visualization - use of matplotlib graphics**


```
import numpy as np
from matplotlib import pyplot as plt

#Generate 100 random numbers 
# https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.randn.html
ys = 10 + np.random.randn(100)
x = [x for x in range(len(ys))]
print("ys = ",ys)
print("x = ",x)

# Plot these numbers
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 150, where=(ys > 150), facecolor='blue', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

# **Plot histogram filled with random numbers**

```
num_bins = 50

z = np.random.randn(1000)

print("Shape of z: ",z.shape," first 5 values:",z[:5])

# the histogram of the data
n, bins, patches = plt.hist(z, num_bins, facecolor='green',  density=True)

plt.title("Histogram Visualization")
plt.xlabel('X-values')
plt.ylabel('Entries')

plt.text(1, .4, r'Random numbers')
plt.xlim(-3, 3)

plt.grid(True)

plt.show()
```

# Working with multiple figures and axes 

MATLAB, and pyplot, have the concept of the current figure and the current axes. All plotting commands apply to the current axes. 
Below is a script to create two subplots.

```
def f1(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def f2(t):
    return np.cos(2*np.pi*t)
   

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

print("t1 = ",t1)

plt.figure()
plt.subplot(211) # 2x1 plots, first plot
plt.plot(t1, f1(t1), color='green', marker='o')
plt.plot(t1, f1(t1), color='black', linestyle='solid')

plt.subplot(212)
plt.plot(t2, f2(t2), color='red', linestyle='dashed')
plt.show()
```
